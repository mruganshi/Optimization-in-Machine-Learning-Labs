```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def g11(x):
    return x[0]**2 + x[1]**2 + 2 * x[2]**2 + x[3]**2 - 5*x[0] - 5*x[1] - 21*x[2] + 7*x[3]

def g12(x):
    return g11(x) + 10 * (x[0]**2 + x[1]**2 + x[2]**2 + x[3]**2 + x[0] - x[1] + x[2] - x[3] - 8)

def g21(x):
    return g11(x) + 10 * (x[0]**2 + 2*x[1]**2 + x[2]**2 + 2*x[3]**2 - x[0] - x[3] - 10)

def g22(x):
    return g11(x) + 10 * (2*x[0]**2 + x[1]**2 + x[2]**2 + 2*x[3] - x[1] - x[3] - 5)

def f1(x):
    return max(g21(x), g22(x))

def f2(x):
    return max(g11(x), g12(x))

def subgrad(x, f1, f2):
    if f1(x) >= f2(x):
        return grad(x, f1)
    else:
        return grad(x, f2)

def grad(x, f):
    h = 1e-5
    gradient = np.zeros_like(x)
    for i in range(len(x)):
        x_plus_h = x.copy()
        x_plus_h[i] += h
        gradient[i] = (f(x_plus_h) - f(x)) / h
    return gradient


```
